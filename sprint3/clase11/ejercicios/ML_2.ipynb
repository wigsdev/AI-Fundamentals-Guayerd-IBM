```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# =========================
# Dataset 
# =========================
data = {
    "edad": [22, 25, 47, 52, 46, 56, 23, 28, 55, 60],
    "salario": [25000, 30000, 52000, 48000, 61000, 72000, 27000, 32000, 69000, 85000],
    "compra": [0, 0, 1, 1, 1, 1, 0, 0, 1, 1]
}

df = pd.DataFrame(data)


X = df[["edad", "salario"]]
y = df["compra"]

# División
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Modelo
modelo = LogisticRegression()
modelo.fit(X_train, y_train)

y_pred = modelo.predict(X_test)

# =========================
# 1. Matriz de Confusión
# =========================
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.title("Matriz de Confusión")
plt.show()


```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# 1) Dataset original (edad / salario)

data = {
    "edad":    [22, 25, 47, 52, 46, 56, 23, 28, 55, 60],
    "salario": [25000, 30000, 52000, 48000, 61000, 72000, 27000, 32000, 69000, 85000]
}

df = pd.DataFrame(data)

X = df[["edad"]]      # variable predictora
y = df["salario"]      # variable a predecir


# 2) División train / test

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 3) Modelo y entrenamiento

modelo = LinearRegression()
modelo.fit(X_train, y_train)


# 4) Predicción

y_pred = modelo.predict(X_test)

# 5) Métricas

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2  = r2_score(y_test, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("R²:", r2)


# 6) Gráfico: datos + recta

plt.figure(figsize=(7,4))
plt.scatter(df["edad"], df["salario"], label="Datos reales")
plt.plot(X, modelo.predict(X), label="Recta de regresión")
plt.xlabel("Edad")
plt.ylabel("Salario")
plt.title("Regresión Lineal: Edad vs Salario")
plt.legend()
plt.show()


# 7) Gráfico: residuos

residuos = y_test - y_pred

plt.figure(figsize=(7,4))
plt.scatter(X_test, residuos)
plt.axhline(0)
plt.xlabel("Edad")
plt.ylabel("Residuos")
plt.title("Gráfico de residuos")
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# 1) Creamos un dataset  (dos características)
X = np.array([
    [1, 1], [1, 2], [2, 1],     # Clase 0
    [8, 8], [8, 9], [9, 8]      # Clase 1
])

y = np.array([0, 0, 0, 1, 1, 1])

# 2) División en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 3) Crear y entrenar el modelo KNN
modelo = KNeighborsClassifier(n_neighbors=3)  # K = 3
modelo.fit(X_train, y_train)

# 4) Predicción
y_pred = modelo.predict(X_test)

# 5) Evaluación
print("Exactitud:", accuracy_score(y_test, y_pred))
print("\nReporte de clasificación:\n", classification_report(y_test, y_pred))
```
