Accuracy


Es el porcentaje de predicciones correctas sobre el total.

```python
from sklearn.metrics import accuracy_score

# Valores reales y predichos
y_real = [1, 0, 1, 1, 0, 1]
y_pred = [1, 0, 1, 0, 0, 1]

# Calcular accuracy
accuracy = accuracy_score(y_real, y_pred)
print("Accuracy:", accuracy)
```

```python
# Accuracy: 0.8333  (≈ 83.3%)
```

Matriz de confusion

Muestra cuántos aciertos y errores hubo en cada clase.
Es clave para problemas de clasificación binaria o multiclase.

```python
from sklearn.metrics import confusion_matrix
import pandas as pd

y_real = [1, 0, 1, 1, 0, 1]
y_pred = [1, 0, 1, 0, 0, 1]

matriz = confusion_matrix(y_real, y_pred)

df_matriz = pd.DataFrame(
    matriz,
    index=["Real: 0", "Real: 1"],
    columns=["Predicho: 0", "Predicho: 1"]
)

print(df_matriz)
```

Error promedio 

MAE Mean Absolute Error, es el promedio de las diferencias absolutas entre las predicciones y los valores reales.
MSE Mean Squared Error, es el promedio de las diferencias al cuadrado entre las predicciones y los valores reales.
RMSE Root Mean Squared Error, es la raíz cuadrada del MSE, da más peso a los errores grandes.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Valores reales y predichos
y_real = [100, 200, 300, 400, 500]
y_pred = [110, 190, 290, 410, 480]

# Calcular métricas
mae = mean_absolute_error(y_real, y_pred)
mse = mean_squared_error(y_real, y_pred)
rmse = np.sqrt(mse)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
MAE: 8.0
MSE: 86.0
RMSE: 9.27
```

| Tipo de problema | Métrica                 | Librería                                    | Ejemplo típico                   |
| ---------------- | ----------------------- | ------------------------------------------- | -------------------------------- |
| Clasificación    | **Accuracy**            | `accuracy_score`                            | Predicciones correctas           |
| Clasificación    | **Matriz de confusión** | `confusion_matrix`                          | Distribución de aciertos/errores |
| Regresión        | **MAE / MSE / RMSE**    | `mean_absolute_error`, `mean_squared_error` | Error medio o cuadrático         |
