# Ejercicio de Visualización: Rendimiento E-commerce

A partir del dataset del ejercicio de la clase 7, se aplicarán distintas visualizaciones para analizar el rendimiento del e-commerce.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo para los gráficos
sns.set_style("whitegrid")

# Datos del ejercicio de la clase 7
data = {
  'Mes':['Ene','Feb','Mar','Abr','May'],
  'Ventas':[45000,52000,38000,61000,48000],
  'Visitantes':[15000,18200,12500,20500,16800],
  'Conversion':[3.2,2.9,3.8,3.1,2.7],
  'Publicidad':[8500,9800,7200,11200,9500],
  'Productos':[450,520,380,610,480]
}

df = pd.DataFrame(data)

print("Dataset a utilizar:")
df.head()
```

### 1. Evolución de Ventas Mensuales (Gráfico de Líneas)
Utilizamos un gráfico de líneas para observar la tendencia de las ventas a lo largo de los meses.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Mes', y='Ventas', data=df, marker='o', sort=False)
plt.title('Evolución de Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas ($)')
plt.show()
```

### 2. Comparación de Eficiencia Publicitaria (Gráfico de Barras)
Calculamos la eficiencia (Ventas / Publicidad) y la comparamos entre los meses usando un gráfico de barras para ver qué mes tuvo mejor retorno.

```python
df['Eficiencia'] = df['Ventas'] / df['Publicidad']

plt.figure(figsize=(10, 6))
sns.barplot(x='Mes', y='Eficiencia', data=df)
plt.title('Eficiencia Publicitaria por Mes (Ventas / Gasto)')
plt.xlabel('Mes')
plt.ylabel('Eficiencia (Ratio Ventas/Gasto)')
plt.show()
```

### 3. Relación Visitantes vs. Ventas (Gráfico de Dispersión)
Un gráfico de dispersión es ideal para visualizar la relación entre dos variables numéricas. Esperamos ver que a más visitantes, más ventas.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Visitantes', y='Ventas', data=df, s=100) # s=100 para hacer los puntos más grandes
plt.title('Relación entre Visitantes y Ventas')
plt.xlabel('Número de Visitantes')
plt.ylabel('Ventas ($)')
plt.show()
```

### 4. Matriz de Correlación (Heatmap)
Finalmente, un heatmap nos permite visualizar la matriz de correlación entre las variables numéricas clave para identificar rápidamente las relaciones más fuertes.

```python
correlation_matrix = df[['Ventas', 'Visitantes', 'Publicidad']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlaciones')
plt.show()
```
